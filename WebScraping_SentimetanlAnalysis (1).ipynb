```python
from bs4 import BeautifulSoup
import requests
```

```python
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
```

```python
url = 'https://blog.goodaudience.com/how-do-i-become-a-data-scientist-f8074232608e'
```

```python
r = requests.get(url)
data = r.text
```

```python
soup = BeautifulSoup(data,'lxml')
```

```python
news_content = soup.find("div",{'class':'section-inner sectionLayout--insetColumn'})

p = news_content.find_all('p')
content = ' '.join(item .text for item in p)
```

```python
content = content.lower()
content
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.style.use('ggplot')
```

```python
Medium = np.array(Image.open('medium.png'))
```

```python
sns.set(rc={'figure.figsize':(16,10)})

wordcloud = WordCloud(mask=Medium,background_color="black").generate(content)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.title('How do I become a data scientist?',size=24)
plt.show()
```
